```python
from extract import Extractor
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Extract from itunes folders, work with copies
extractor = Extractor()
df_apple, df_loaded = extractor.process_itunes_tracks()
df_apple_wip = df_apple.copy()
df_music_wip = df_loaded.copy()
```

```python
# Drop columns df_music_wip
df_music_wip = df_music_wip.drop(columns=[
    "Encoded with", "BPM", "Part of Gapless Album", "Cover Art pieces", "Copyright", "Content Rating", "Media Type",
    'CRî" 1184108314 vs 7910639', "CRî is suspect", "iTunes Account", "Purchase Date", "ReadAtom", "iTunes Account Type",
    "Comments", "TV Episode", "TV Season", "Sort Composer", "TV Show"
    ])

# Drop columns df_apple_wip
df_apple_wip = df_apple_wip.drop(columns=[
    "Copyright", "Content Rating", "Media Type", "iTunes Account", "Cover Art pieces", "ReadAtom"
    ])

# Drop columns
df_music_wip = df_music_wip.drop(columns=["Sort Name", "Sort Artist", "Sort Album"])

# Rename columns
df_music_wip = df_music_wip.rename(columns={"Name": "Sort Name", "Artist": "Sort Artist", "Album": "Sort Album"})

# Combine into a single dataframe
df_combined = pd.concat([df_music_wip, df_apple_wip])

# Left strip and Right strip all the strings
df_obj = df_combined.select_dtypes(['object'])
df_obj[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
df_combined = df_obj
```
